# Comparing with structural similarity index (SSIM)
SSIM from scikit-image will do the job for us.

```python
import cv2
import numpy as np
from skimage.measure import compare_ssim
from skimage.transform import resize
from matplotlib import pyplot as plt
```

```python
# Find the paths of parle g from both
test_path = '../../data/test_images/test_image_1.jpeg'
path = '../../data/Biscuits/400018.jpg'
```

```python
# View both images
test_img = cv2.imread(test_path, 0)
img = cv2.imread(path, 0)

fig = plt.figure()
fig.add_subplot(121)
plt.imshow(img, cmap='gray')
fig.add_subplot(122)
plt.imshow(test_img, cmap='gray')
plt.tight_layout()
plt.show()
```

```python
# Structural similarity index
# Resize the image to 
test_img = resize(test_img, output_shape=img.shape)

# SSIM
s = compare_ssim(test_img.astype('float'), img.astype('float'))
print("Similarity: {:.5f}".format(s))
```
