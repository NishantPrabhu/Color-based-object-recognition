# Colorspace search with CNN features
Using the colors extracted by the previous algorithm and features from a pretrained network, let's see if we can narrow down on the right product.

```python
import os
import cv2
import pickle
import numpy as np 
from glob import glob
from ntpath import basename
from tqdm.notebook import tqdm
from sklearn.cluster import MiniBatchKMeans
from matplotlib import pyplot as plt
from IPython.display import clear_output

import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.applications import InceptionV3
from tensorflow.keras.applications.inception_v3 import preprocess_input
from tensorflow.keras.preprocessing.image import load_img, img_to_array
```

```python
# Load model
# Remove top layer

inception_model = InceptionV3(weights='imagenet')
inception_model = Model(inception_model.inputs, inception_model.layers[-2].output)
```

```python
# Feature extraction pipeline

def extract_cnn_features(root_dir, model):
    folders = os.listdir(root_dir)
    progress = tqdm(total=len(folders), position=0, desc='Progress')
    folder_name = tqdm(total=0, position=1, bar_format='{desc}')
    file_feature_map = {}
    
    for folder in folders:
        folder_name.set_description_str(f"Now processing {folder} ...")
        files = glob(root_dir+'/'+folder+'/*.jpg')
        folder_progress = tqdm(total=len(files), position=2, desc='Folder', leave=False)
        for path in files:
            img = load_img(path, target_size=(299, 299))
            img = img_to_array(img)
            img = np.expand_dims(img, axis=0)
            x = preprocess_input(img)
            features = model(x)
            file_feature_map.update({path: features.numpy()})
            folder_progress.update(1)
        progress.update(1)
            
    return file_feature_map
```

```python
# Run the above function

file_feature_map = extract_cnn_features("../../data", inception_model)
```

```python
# Generate file_index_map and feature stack

file_index_map = {k: i for i, k in enumerate(file_feature_map.keys())}
feature_stack = np.asarray(list(file_feature_map.values())).squeeze(1)
```

```python
# Save the files

with open("../../saved_data/20 Jun/file_feature_map.pkl", "wb") as f:
    pickle.dump(file_feature_map, f)
    
with open("../../saved_data/20 Jun/file_index_map.pkl", "wb") as f:
    pickle.dump(file_index_map, f)
    
with open("../../saved_data/20 Jun/feature_stack.pkl", "wb") as f:
    pickle.dump(feature_stack, f)
```

# Algorithm
Find candidate matches using jaccard score of labels, then find final match by closeness of inception features.

```python
# Load objects
root = '../../saved_data/20 Jun/'

with open(root+'all_colors.pkl', 'rb') as f:
    all_colors = pickle.load(f)
    
with open(root+'feature_stack.pkl', 'rb') as f:
    feature_stack = pickle.load(f)
    
with open(root+'file_color_map.pkl', 'rb') as f:
    file_color_map = pickle.load(f)
    
with open(root+'file_feature_map.pkl', 'rb') as f:
    file_feature_map = pickle.load(f)
    
with open(root+'file_index_map.pkl', 'rb') as f:
    file_index_map = pickle.load(f)
    
with open(root+'kmeans_clt.pkl', 'rb') as f:
    kmeans_clt = pickle.load(f)
```

```python
class ProductFinder(object):
    
    def __init__(self, n_colors, kmeans_clt, file_index_map, 
                 feature_stack, all_colors, file_color_map, top_k=3):
        self.n_colors = n_colors
        self.kmeans_clt = kmeans_clt
        self.file_index_map = file_index_map
        self.feature_stack = feature_stack
        self.all_colors = all_colors
        self.file_color_map = file_color_map
        self.top_k = top_k
        
    def jaccard_score(self, list_1, list_2):
        union = set(list_1).union(list_2)
        intr = set(list_1).intersection(list_2)
        return len(intr)/len(union)
        
    def find_candidates(self, img):
        img = img.reshape(img.shape[0]*img.shape[1], -1)
        mb_kmeans = MiniBatchKMeans(n_clusters=self.n_colors)
        mb_kmeans.fit(img)
        labels = self.kmeans_clt.predict(mb_kmeans.cluster_centers_)
        js = [self.jaccard_score(labels, v) for v in self.file_color_map.values()]
        N = int(0.01*len(js))
        candidates = sorted(range(len(js)), key=lambda i: js[i])[-N:]  
        candidate_paths = np.array(list(self.file_color_map.keys()))[candidates].tolist()
        return candidates, candidate_paths
    
    def find_feature_match(self, img, img_path, model, candidates, candidate_paths):
        img = load_img(img_path, target_size=(299, 299))
        img = img_to_array(img)
        img = np.expand_dims(img, axis=0)
        x = preprocess_input(img)
        img_fs = model(x)
        
        dist = [np.linalg.norm(img_fs-self.feature_stack[i]) for i in candidates]
        winners = sorted(range(len(dist)), key=lambda i: dist[i])[:self.top_k]
        winner_paths = [candidate_paths[i] for i in winners]
        return winners, winner_paths
    
    def evaluate_model(self, root_dir, num_examples=500):
        folders = os.listdir(root_dir)
        progress = tqdm(total=len(folders), position=0, desc='Progress')
        folder_name = tqdm(total=0, position=1, bar_format='{desc}')
        correct = 0
        top3_correct = 0
        total_files = 0
        
        for folder in folders:
            folder_name.set_description_str(f"Now processing {folder} ...")
            files = glob(root_dir+'/'+folder+'/*.jpg')
            folder_progress = tqdm(total=len(files), position=2, desc='Folder', leave=False)
            for path in files:
                total_files += 1
                img = cv2.imread(path)
                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                cnd, cnd_paths = self.find_candidates(img)
                wins, win_paths = self.find_feature_match(img, path, inception_model, 
                                                          cnd, cnd_paths)
                win_names = [basename(path) for path in win_paths]
                if win_names[0] == basename(path):
                    correct += 1
                if basename(path) in win_names:
                    top3_correct += 1
                folder_progress.update(1)
                # Check break condition
                if total_files == num_examples:
                    break
            progress.update(1)
        
        clear_output()
        print("Accuracy: {:.2f}%".format(100.*correct/total_files))
        print("Top 3 accuracy: {:.2f}%".format(100.*top3_correct/total_files))
        return correct/total_files, top3_correct/total_files 
    
    def show_example(self):
        filenames = list(self.file_color_map.keys())
        i = np.random.choice(np.arange(0, len(filenames)), size=1)[0]
        path = filenames[i]
        
        img = cv2.imread(path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        cnd, cnd_paths = self.find_candidates(img)
        wins, win_paths = self.find_feature_match(img, path, inception_model, cnd, cnd_paths)
        
        # Show the image and matches
        matched_imgs = [cv2.imread(path) for path in win_paths]
        matched_imgs = [cv2.cvtColor(i, cv2.COLOR_BGR2RGB) for i in matched_imgs]
        
        print("Test image ...")
        plt.imshow(img)
        plt.show()
        print(f"\nTop {self.top_k} matched images ...")
        fig = plt.figure(figsize=(10, 5))
        for i in range(self.top_k):
            fig.add_subplot(1, self.top_k, i+1)
            plt.imshow(matched_imgs[i])
        plt.tight_layout()
        plt.show()
```

```python
finder = ProductFinder(
    n_colors=10,
    kmeans_clt=kmeans_clt,
    file_index_map=file_index_map,
    feature_stack=feature_stack,
    all_colors=all_colors,
    file_color_map=file_color_map,
    top_k=3
)

# Evaluate model
finder.evaluate_model(root_dir='../../data', num_examples=500)
```
